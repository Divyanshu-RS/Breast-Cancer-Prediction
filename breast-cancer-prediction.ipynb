## **Project Overview: Breast Cancer Prediction Using Machine Learning**
This project focuses on predicting breast cancer through comprehensive data analysis and machine learning. The key tasks include data preprocessing, visualization of feature distributions, model evaluation, and performance comparison using cross-validation techniques.

### **Dataset Details:**
- Dataset: Breast Cancer Wisconsin (Diagnostic) dataset
- Instances: 569
- Features: 30 real-valued features including radius, texture, perimeter, area, smoothness, compactness, concavity, concave points, symmetry, and fractal dimension, with mean, standard error, and worst (largest) values recorded for each.

### **Key Tasks:**
- Data Preprocessing: Cleaning and preparing the dataset for analysis.
- Feature Visualization: Analyzing and visualizing the distributions of features to understand their characteristics.
- Model Evaluation: Training various machine learning models and evaluating their performance using cross-validation to ensure robustness.
- Performance Comparison: Comparing the effectiveness of different models to determine the best-performing algorithm for breast cancer prediction.

### **Classification Task:**

- Classes: Malignant (212 cases) and Benign (357 cases)
- Objective: Develop a reliable predictive model to assist in the early detection and diagnosis of breast cancer, thereby enhancing patient outcomes.

  
The dataset is publicly available through the UCI Machine Learning Repository or the University of Wisconsin-Madison's CS FTP server.

## **Prediction with Cross Validation**

In this analysis, we evaluate the performance of a model using cross-validation (CV) to compare the accuracy scores between two datasets: one with all features and one with highly correlated features removed. This helps us understand the impact of feature selection on model performance.

#### **Function Definition:**
- **evaluate_cv(data, model):** This function performs cross-validation to assess the model's accuracy.
  
#### **Cross-Validation Setup:**
- **KFold:** Defines a 5-fold cross-validation strategy, shuffling the data and setting a random seed for reproducibility.
  
#### **Evaluation Loop:**
- **Splitting Data:** For each fold, the data is split into training and testing sets. Features (excluding the first column) are used for training, and the target variable (first column) is used for predictions.
- **Model Training:** The model is trained on the training set.
- **Model Evaluation:** The accuracy of the model is calculated on the testing set and stored in the scores list.
  
### **Results:**
The average accuracy across all folds is calculated and printed to evaluate model performance.

#### **Impact:**
- **Model Comparison:** By comparing the accuracy scores of the dataset with all features and the one with correlated features removed, we can determine the effect of feature selection on the model's performance.
- **Feature Selection Effectiveness:** Helps in understanding whether removing highly correlated features improves or degrades the model's predictive accuracy.

```python
from sklearn.model_selection import KFold

def evaluate_cv(data, model):
    # Define the number of folds for cross-validation
    kfold = KFold(n_splits=5, shuffle=True, random_state=42)

    # Lists to store evaluation metrics (e.g., accuracy)
    scores = []
    count = 1

    for train_index, test_index in kfold.split(data):
    
        # Split data into training and testing sets for this fold
        X_train, X_test = data.iloc[train_index, 1:], data.iloc[test_index, 1:]  # Select features (all except first column)
        y_train, y_test = data.iloc[train_index, 0], data.iloc[test_index, 0]  # Select target variable (first column)

        # Train the model on the training set
        model.fit(X_train, y_train)

        # Evaluate the model on the testing set (e.g., calculate accuracy)
        score = model.score(X_test, y_test)
        scores.append(score)
    print(f'Accuracy: {np.mean(scores)}')
    return np.mean(scores)
```

```python
from sklearn.svm import SVC

model = SVC()
svc_df = evaluate_cv(df, model)
svc_data = evaluate_cv(data, model)
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
dtc_df = evaluate_cv(df, model)
dtc_data = evaluate_cv(data, model)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100,min_samples_split=25, max_depth=7, max_features=2)
rfc_df = evaluate_cv(df, model)
rfc_data = evaluate_cv(data, model)
```

### **K-Nearest Neighbors (KNN)**

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier()
knn_df = evaluate_cv(df, model)
knn_data = evaluate_cv(data, model)
```

### **Gradient Boosting Classifier (GBC)**

```python
from sklearn.ensemble import GradientBoostingClassifier

model = GradientBoostingClassifier()
gbc_df = evaluate_cv(df, model)
gbc_data = evaluate_cv(data, model)
```

### **AdaBoost Classifier (ABC)**

```python
from sklearn.ensemble import AdaBoostClassifier

model = AdaBoostClassifier()
abc_df = evaluate_cv(df, model)
abc_data = evaluate_cv(data, model)
```

# **Model Performance Comparison**

This code snippet visualizes the performance scores of various machine learning models using a horizontal bar plot. The goal is to compare the performance of models on two different datasets: one with all features and one with highly correlated features removed.

### **Scores Dictionary:**
scores dictionary holds the performance scores of different models (LR, SVC, GNB, DTC, RFC, KNN, GBC, ABC) on both datasets (with features intact [data] and with selected features [df]).

### **Plot Preparation:**
- **Labels and Values:** Extracted from the scores dictionary to use in the plot.
- **Colors:** Alternating colors (blue and red) are defined to differentiate between the two datasets.

### **Plotting:**
- A horizontal bar plot is created with plt.barh(), where bars represent the performance scores.
- Labels and Titles: The x-axis is labeled 'Scores', the y-axis is labeled 'Models', and the plot is titled 'Model Performance Scores'.
- X-axis Limit: Set between 0.87 and 0.97 to focus on the range of scores.
  
### **Legend:**
- A legend is added to distinguish between 'Whole Data' and 'Clean Data'.

### **Display:**
- The plot is displayed using plt.show(), providing a clear comparison of model performances on different datasets.

This visualization helps in quickly assessing how each model's performance varies when using the full dataset versus a cleaned dataset with selected features.

```python
# defining the scores dictionary
scores = {
    'LR [df]': lr_df, 'LR [data]': lr_data, 
    'SVC [df]': svc_df, 'SVC [data]': svc_data, 
    'GNB [df]': gnb_df, 'GNB [data]': gnb_data, 
    'DTC [df]': dtc_df, 'DTC [data]': dtc_data, 
    'RFC [df]': rfc_df, 'RFC [data]': rfc_data,
    'KNN [df]': knn_df, 'KNN [data]': knn_data,
    'GBC [df]': gbc_df, 'GBC [data]': gbc_data,
    'ABC [df]': abc_df, 'ABC [data]': abc_data
}

# Extract labels and scores
labels = list(scores.keys())
values = list(scores.values())

# Define alternating colors (blue and red)
colors = ['blue', 'red'] * (len(labels) // 2)

# Create the horizontal bar plot
plt.figure(figsize=(10, 6))
bars = plt.barh(labels, values, color=colors)
plt.xlabel('Scores')
plt.ylabel('Models')
plt.title('Model Performance Scores')
plt.xlim([0.87,0.97])
plt.tight_layout()

# Add a legend
plt.legend(bars, ['Whole Data', 'Clean Data'])

# Show the plot
plt.show()
```
